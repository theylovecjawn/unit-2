```python
import matplotlib.pyplot as plt

y = [1,2,3,4]
x = [22,87,0,18]

plt.plot(x,y)
plt.ylabel('Kills')
plt.xlabel('People')
plt.title("Kill Per Person Ratio")
```

```python
pie_percents = [30,10,25,100] 
#This is me setting the percents in the pie chart
pie_labels = ["Ash", "Twitch", "Sledge", "Osa"]
# Sets the different labels for the pie chart
explode = [0, 0, 0, 0.2] 
#creation of the explode to push it apart form the pie chart
plt.pie(y, labels = pie_labels, autopct='%1.1f%%', explode = explode) 
#creation of the pie chart using labels and adding the percentages to it, and use of the explode
plt.title("Best Operators in R6") #sets the title
```
