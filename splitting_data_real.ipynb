*Use the code block below to open `sales.csv` with python. Then create a variable called `content` which is equal to a list of each line of the file*

````python
with open('sales.csv', 'r') as file:
    #reads the content of the file
    content = file.readlines() #prints the lines

print(content) #prints the content  
````

*Using the `split` function and a `for` loop calculate the total sales in a month*

*Hint: you will need a seperate variable for each month*

````python
jan = 0 #creates variable for jan
feb = 0#creates variabel for feb
mar = 0 #creates variabel for mar
del content[0]
for line in content:
    parts = line.strip().split(',')
    
    #making sure that the date is in the format MM/DD/YYYY
    date = parts[3]  # Use index 0 to access the date
    month = date.split('/')[0]
    total = float(parts[2])
    
    if month == "1":
        jan += total  #adds the sales value to the corresponding month
    elif month == "2":
        feb += total
    elif month == "3":
        mar += total

print("January total sales:","$", jan) #prints  the total sales for jan
print("February total sales:","$", feb) #prints the total sales for jan
print("March total sales:","$",mar) #prints the total sales for jan
````

*Using your calculated total sales create a graph to represent this data and write a written analysis*

````python
import matplotlib.pyplot as plt  #imports the Pyplot from the matplotlib 
y = [jan, feb, mar] #list for the Y axis
x = ["Janurary", "Feburary", "March"] #list for x axis


plt.plot(x, y) #plots the graph
plt.ylim(ymin=0) #sets the y axis minimum to 0
plt.xlabel("Months") #names x axis to Months
plt.ylabel("Total Amount") #y axis name is now total
plt.title("Sales Per First three months") #sets the titles to sales per first three months
plt.grid(True, linestyle='--', alpha=0.6) #adds grids to the graph
````

## After Analyzing the graph I see that after the amount gained in Janurary the the amoutn made dropped almost $5,000. Then going into March it increased a substantial amount surpassing that of the price for Janurary.

*Using the `split` function, seperate each row of data. Then append the customer's name to a list called `customers`*

````python
customers = [] #creates the customers list
for row in content:
    #splits each row by a comma to separate the elements
    piece = row.split(',')
    
    
    #checks if there is at least one element (in case there are empty rows)
    if piece:
        #appends the first element (customer's name) to the customers list
        customers.append(piece[0])

print(customers) #prints the names
````

*Using the `Counter` from the `collections` library, and your list of customers create a graph to represent this data and write a written analysis*

````python
customers_count = Counter(customers) #Uses the Counter from the collections

keys = customers_count.keys() #gets the keys from the dictionary
values = customers_count.values()# gets the values from the dictionary

plt.figure(figsize=(12, 6))  #adjusts the figure size
plt.bar(keys, values, color='g', label='Frequency Data') #creates the bar and adds the x,y axis and changed the color and label
plt.xlabel('Customer Names', fontsize=8) # labels the name of the x "State Names"
plt.ylabel('Frequency', fontsize=12)# labels the name of the x "State Names"
plt.title("Frequency Of People's Purchase", fontsize=20) # the title is created
plt.xticks(rotation=70)  # rotates the x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=.5)  # add a grid for the y-axis
plt.legend() #creates the bar graph legend 
````
