### 1. What is the percentage amount of males who survived?

```python
#importings pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

#reads the Titanic dataset into a DataFrame
df = pd.read_csv("titanic.csv")
#filters for males only
male_df = df[df['Sex'] == 'male']
#counts the number of survivors and non-survivors among males
survival_counts = male_df['Survived'].value_counts()
#defines labels and sizes for the pie chart
labels = ['Non-Survived', 'Survived']
sizes = survival_counts[0], survival_counts[1]
#defines the 'explode' parameter to emphasize a slice of the pie (in this case, 'Survived')
explode = (0, 0.1)
#creates a pie chart
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', startangle=140)
plt.axis('equal') #equal aspect ratio ensures that the pie chart is circular
#sets the title of the pie chart
plt.title('Survival Percentage for Males')
#shows the pie chart
plt.show()
```
