``` 1. What is the percentage amount of males who survived?```

````python
#importings pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

#reads the Titanic dataset into a DataFrame
df = pd.read_csv("titanic.csv")
#filters for males only
male_df = df[df['Sex'] == 'male']
#counts the number of survivors and non-survivors among males
survival_counts = male_df['Survived'].value_counts()
#defines labels and sizes for the pie chart
labels = ['Non-Survived', 'Survived']
sizes = survival_counts[0], survival_counts[1]
#defines the 'explode' parameter to emphasize a slice of the pie (in this case, 'Survived')
explode = (0, 0.1)
#creates a pie chart
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', startangle=140)
plt.axis('equal') #equal aspect ratio ensures that the pie chart is circular
#sets the title of the pie chart
plt.title('Survival Percentage for Males')
#shows the pie chart
plt.show()
````

``` 2. What is the average amount of money males paid in fairs that survived versus the ones that died?```

````python

x = ["Died","Survived"]
y = [male_df[male_df['Survived'] == 0]['Fare'].mean() , male_df[male_df['Survived'] == 1]['Fare'].mean()]
print(y)
plt.bar(x, y)
plt.xlabel('Death retention', fontsize=8) # labels the name of the x "State Names"
plt.ylabel('Average of Fares', fontsize=12)# labels the name of the x "State Names"
plt.title("The average of the People that Died/ Survived", fontsize=20) # the title is created
plt.xticks(rotation=70)  # rotates the x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=.5)  # add a grid for the y-axis
plt.legend() #creates the bar graph legend 
````

``` 3. What percent of Males had at least one other relative with them on the trip?```

````python
#imports the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

#reads the Titanic dataset into a DataFrame
df = pd.read_csv("titanic.csv")

#filters for males only
male_df = df[df['Sex'] == 'male']

#filters males with Parch > 0 and males with Parch == 0
male_with_parch_df = male_df[male_df['Parch'] > 0]
male_without_parch_df = male_df[male_df['Parch'] == 0]

#calculates the average fare for each group
average_fare_with_parch = male_with_parch_df['Fare'].mean()
average_fare_without_parch = male_without_parch_df['Fare'].mean()

#creates data for the bar chart
x = ["With Parch", "Without Parch"]
y = [average_fare_with_parch, average_fare_without_parch]

#creates a bar chart to visualize the average fare
plt.bar(x, y, color='blue')
plt.xlabel('Passenger group')
plt.ylabel('Average fare')
plt.title('Average fare for males with Parch > 0')
plt.show()
````
