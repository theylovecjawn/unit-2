Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt 

f = open('states.txt','r')
NY = 0
FL = 0
MO = 0
TX = 0
KS = 0
for row in f: 
    row = row.strip()
    if row == "NY":
        NY += 1 
    elif row == "MO":
        MO += 1
    elif row == "FL":
        FL += 1
    elif row == "TX":
        TX += 1
    elif row == "KS":
        KS += 1

plt.bar(["NY", "FL", "MO", "TX", "KS"], [NY, FL, MO, TX, KS], color = 'g', label = 'States Data')
plt.xlabel('State Names', fontsize = 12) 
plt.ylabel('Amount', fontsize = 12) 
plt.title('State Amount', fontsize = 20) 
plt.legend() 
plt.show() 
````
